### Desafio 02: Adicionar uma coluna mostrando a região (Norte, Nodeste, Sul, Suldeste e Centro-Oeste) de cada estado.

```python
#Criando lista de regiao pelo indice do dataframe
lst = []
for n in range(0,27):
    if usaveis.index[n][0] == '1':
        lst.append('Norte')
    elif usaveis.index[n][0] == '2':
        lst.append('Nordeste')
    elif usaveis.index[n][0] == '3':
        lst.append('Sudeste')
    elif usaveis.index[n][0] == '4':
        lst.append('Sul')
    elif usaveis.index[n][0] == '5':
        lst.append('Centro Oeste')

#add lista no dataframe
usaveis['Regiao'] = lst
usaveis.head(8)
```

### Desafio 03: Formatar o gráfico de custos por mês dos 5 estados, deixando ele agradável (Bonitão, segundo o Gui)

```python
usaveis['Total'] = usaveis.sum(axis = 1)
five_uf = usaveis.sort_values('Total', ascending = False).head(5)
five_uf = five_uf.drop('Total', axis=1)
five_uf = five_uf.drop('Regiao', axis=1)
```
