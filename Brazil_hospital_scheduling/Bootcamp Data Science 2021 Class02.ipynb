```python
# Dados numericos apenas como lista. faltou o index!!!
colunas_usaveis = df.mean().tolist()
colunas_usaveis
```

```python
# Falta o estado, para isso vamos chamar o index.tolist() e inserir os estados com insert
colunas_usaveis = df.mean().index.tolist()
colunas_usaveis.insert(0, 'Unidade da Federação')
type(colunas_usaveis), colunas_usaveis
```

```python
# Fazer um slicing no dataframe df
cols_use = df[colunas_usaveis]
cols_use.head()
```

```python
# Colocar estados como indice.
usaveis = cols_use.set_index('Unidade da Federação')
usaveis.head()
```
